```python
newdata=pd.read_csv("Downloads/student_marks.csv")
print(newdata)
```

```python
student=pd.read_csv("Downloads/students_20 (1).csv")
print(student)
```

```python
#Show the first 5 rows of the DataFrame.
student.head()
```

```python
list(student.columns)
```

```python
#student.rename(columns={'Student Name':'Name'})
```

```python
#Show only the "Student Name" and "Math" columns.
student[['Student Name','Math']]
```

```python
#Find the average marks in each subject
student[['Math','Science','English']].mean()
```

```python
#Which student scored the highest in Science?
maxmark=student['Science'].max()
print("Highest Mark is",maxmark)
maxmarkstud=student[student['Science']==maxmark]
print("Highest Mark scored student ")
print(maxmarkstud)
```

```python
specific=student.loc[student['Science']==maxmark,"Student Name"]
print(specific)
```

```python
#Select students who scored more than 90 in English.

morethan=student[student['English']>90]
print(morethan)
```

```python
names=student.loc[student['English']>90,"Student Name"]
print(names)
```

```python
newnames=student.loc[student['English']>90,["Student Name","English"]]
print(newnames)
```

```python
#Sort students by Math marks in descending order.
order=student.sort_values('Math',ascending=False)[["Student Name","Math"]]
print(order)
```

```python
desc=student['Math'].sort_values(ascending=False),["Student Name"]
print(desc)
```

```python
#Add a new column "Total" = Math + Science + English.
student["Total"]=student["Math"]+student["Science"]+student["English"]
print(student)
```

```python
#Find the student with the highest total marks.
highestmark=student["Total"].max()
print(highestmark)
highest=student.loc[student["Total"]==highestmark,["Student Name"]]
print(highest)
```

```python
#Show students who scored less than 80 in at least one subject.
least=student[(student['Math']<80) | (student['Science']<80) | (student['English']<80)]
print(least)
```

```python
#What is the highest and lowest mark in Math?
print("Highest Mark in Maths is ",student['Math'].max())
print("Lowest Mark in Maths is",student['Math'].min())
```

```python
#Save the updated DataFrame (with totals) to a new CSV.
student.to_csv("Student_Marks_with_total.csv",index=False)
print(student)
```

```python
student.drop(columns=["Total"]).to_csv("Student_Marks_with_total.csv",index=False)
print(student)
```

```python
#Find the average total marks of all students.
avg=student["Total"].mean()
print(avg)
```

```python
#Show only the top 3 students in English.
top3=student.sort_values('English',ascending=False).head(3)
print(top3)
```

```python
#Count how many students scored above 85 in Science.
count=(student['Science']>85).sum()
print(count)
```

```python
#Create a new column "Grade" based on Total Marks

student["Grade"]=pd.cut(student["Total"],bins=[0,239,257,300],labels=['C','B','A'])
print(student[["Student Name","Total","Grade"]])
```

```python
#Show the average marks per subject for each Grade.

st=student.groupby("Grade")[["Math","Science","English"]].mean()
print(st)
```

```python
#Show students who have the same marks in any two subjects.
student[(student["Math"]==student["English"]) | (student["Math"]==student["Science"]) | (student["Science"]==student["English"])]
```

```python
#Show the top 5 students based on total marks only (Student Name + Total).
top5=student[["Total","Student Name"]].sort_values('Total',ascending=False).head(5)
print(top5)
```

```python
#Save only Student Name and Grade to a new CSV.
student[["Student Name","Grade"]].to_csv("student_grade.csv")
```

```python
newsales=pd.read_csv("Downloads/company_sales_person.csv")
print(newsales)
```

```python
#3.Find the shape (rows × columns) of the dataset.
newsales.shape
```

```python
#4.Display all column names and their data types.
list(newsales.columns)
newsales.dtypes
```

```python
#5.Check if there are any missing values
print(newsales.isnull().sum())
```

```python
#6.Display only the SalesPerson and TotalSales columns
newsales[["SalesPerson","TotalSales"]]
```

```python
#7.Show all sales records from the South region.
newsales[newsales["Region"]=="South"]
```

```python
#8.Find all sales where Units > 10.
newsales[newsales["Units"]>10]
```

```python
#9.Show sales records for the product Laptop.
newsales[newsales["Product"]=="Laptop"]
```

```python
#10.Display all sales made by Bob.
newsales[newsales["SalesPerson"]=="Bob"]
```

```python
#11.Add a new column DiscountedSales = TotalSales × 0.9 (10% discount).
newsales["Discountedsales"]=newsales["TotalSales"]*0.9
print(newsales)
```

```python
#12.Add a new column Profit = 25% of TotalSales.
newsales["Profit"]=newsales["TotalSales"]*0.25
print(newsales)
```

```python
#13.Sort the sales by TotalSales in descending order.
newsales.sort_values("TotalSales",ascending=False)
```

```python
#14.Replace the region “North” with “North Zone”.
newsales["Region"]=newsales["Region"].replace("North","North Zone")
newsales
```

```python
#15.Convert the Date column into month names.
newsales["Date"]=pd.to_datetime(newsales["Date"])
newsales["Month"]=newsales["Date"].dt.month_name()
print(newsales)
```

```python
#16.Find the total sales per Region.
print(newsales.groupby("Region")["TotalSales"].sum())
```

```python
#17.Find the average Units sold per Product.
print(newsales.groupby("Product")["Units"].mean())
```

```python
#18.Count how many sales each SalesPerson made.
print(newsales["SalesPerson"].value_counts())
```

```python
#19.Find the highest TotalSales for each Region.
print(newsales.groupby("Region")["TotalSales"].max())
```

```python
#20.Find the total revenue per Product.
print(newsales.groupby("Product")["TotalSales"].sum())
```

```python
#21.Identify the top-performing SalesPerson in terms of revenue.
print(newsales.groupby("SalesPerson")["TotalSales"].sum().sort_values(ascending=False).head(1))
```

```python
#22.Find the product that generated the highest overall sales.
print(newsales.groupby("Product")["TotalSales"].sum().sort_values(ascending=False).head(1))
```

```python
#23.Show the monthly total sales trend.
newsales.groupby("Month")["TotalSales"].sum()
```

```python
#24.Show sales where two different products have the same UnitPrice.
print(newsales.groupby("UnitPrice")["Product"].unique())
```
