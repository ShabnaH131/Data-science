```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,3,4,5]
plt.plot(x,y)
plt.show()
```

```python
x=[2,3,4]
y=[3,4,5]
plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Distance")
plt.title("Distance over Time")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y,color="red",linestyle="--",marker="o")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.title("Distance over Time")
plt.show()
```

```python
x=[8,10,12,14]
y=[3,8,2,1]
plt.figure(figsize=(7,6))
plt.plot(x,y,color="gold",linestyle="-",marker="^")
plt.xlabel("speed (km/hr)")
plt.ylabel("Time (hr)")
plt.title("Distance Covered")
plt.show()
```

```python
#bar chart
fruits=["Apple","Orange","Grapes"]
sales=[10,12,5]
colours=["brown","gold","green"]
plt.bar(fruits,sales,color=colours)
plt.show()
```

```python
#3. pie Chart
num=[20,30,40,10]
name=["A","B","C","D"]
plt.pie(num,labels=name,autopct="%1.1f%%")
plt.show()
```

```python
colours=["#FFFFCB","#F0E68C","#AAA662","#808000"]
plt.pie(num,labels=name,colors=colours,autopct="%1.1f%%")
plt.title("Sales")
plt.show()
```

```python
import seaborn as sns
sizes=[10,20,30,40]
name=["Stationary Items","Food Items","Decor Items","Others"]
colour=sns.color_palette('viridis',len(sizes))
plt.pie(sizes,labels=name,colors=colour,autopct="%1.1f%%",startangle=90)
plt.show()
                         
```

```python
percentages=[17,34,25,12,12]
category=["Doctors","Engineers","IT Professionals","Others","Business"]
colours=sns.color_palette('pastel', len(percentages))
plt.pie(percentages, labels=category,colors=colours, autopct="%1.1f%%")
plt.title("Categories")
plt.show()
```

```python
#4. Scatter plot
x=[78,9,5,56]
y=[5,86,56,90]
colours=["red","green","blue","gold"]
plt.scatter(x,y,color=colours)
plt.show()
```

```python
#5. horizontal bar graph
fruits=["Apple","orange","Grapes","Banana"]
size=[20,40,10,30]
colours=["#FFFFCB","#F0E68C","#AAA662","olivedrab"]
plt.barh(fruits,size,color=colours)
plt.show()
```

```python
#6. Histogram
data=[90,67,8,34,9,23,45,76,12,48,37,29,68]
plt.hist(data, bins=5, color="#FFFFCB",edgecolor="skyblue")
plt.show()
```

```python
data2=[2,39,63,61,9,1,8]
plt.hist(data2, bins=3, color="#F0E68C",edgecolor="GREEN")
plt.show()
```

```python
DATA3=[12,5,50,56,60,80,89,90,100]
plt.hist(DATA3, bins=5, color="olivedrab",edgecolor="gold")
plt.show()
```

```python
#7. Box Plot
datas=[1,6,5,7,8,9,0,3,4]
plt.boxplot(datas)
plt.title("Box Plot")
plt.show()
```

```python
newdata=[67,89,45,39,10]
plt.boxplot(newdata)
plt.show()
```

```python
#8. Stem Plot
x=[16,4,6,8]
y=[1,4,7,9]
plt.stem(x,y)
plt.title("Stem Plot")
plt.show()
```
