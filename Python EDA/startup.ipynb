## Project Name - Startup Funding
### Project Type - EDA (Exploratory Data Analysis)



Startup funding EDA is the process of exploring, cleaning, and analyzing data about startups and their funding patterns. The goal is to uncover trends, insights, and patterns in how startups receive investments, which industries attract the most money, and who the most active investors are.

## 1. Importing Libraries

```python
df.drop('Remarks',axis=1,inplace=True)
#df.drop('Amount in USD',axis=1,inplace=True)
df.drop('SubVertical',axis=1,inplace=True)
df['Industry Vertical']=df['Industry Vertical'].fillna('Unknown')
df['City  Location']=df['City  Location'].fillna('Unknown')
df['Investors Name']=df['Investors Name'].fillna('Unknown')
df['InvestmentnType']=df['InvestmentnType'].fillna('Unknown')
df['Amount in USD'] = pd.to_numeric(df['Amount in USD'], errors='coerce')
df['Amount in USD']=df['Amount in USD'].fillna(df['Amount in USD'].mean())
```

```python
df.isnull().sum()
```

```python
df.duplicated().value_counts()
```

```python
df.columns = df.columns.str.replace(' ', '')
df
```

## Findings

## Top 5 Highest number of Startups in India

```python
df['CityLocation'] = df['CityLocation'].str.strip().str.lower().replace({
    'Bangalore': 'bengaluru',
    'Bengaluru': 'bengaluru'
})
```

```python
high=df.groupby('CityLocation')['StartupName'].nunique().sort_values(ascending=False)
city=high.head(5)
print("Highest Number of Startups in India\n",city)
```

```python
city.plot(kind='bar',color='yellow')
plt.show()
```

### Conclusion :
The dataset reveals that **Bengaluru** hosts the **highest** number of startups in India. This concentration highlights the city’s role as a breeding ground for innovation, supported by favorable infrastructure, skilled talent, and a strong investor network.

### List of Top 5 Investors

```python
top_investors = df['InvestorsName'].value_counts().head(5)
top_investors
```

```python
top_investors.plot(kind='bar',color='skyblue')
plt.show()
```

### Conclusion : 

The data shows that **Undisclosed Investors** have made the **highest** number of investments, followed by Ratan Tata, Unknown, and Indian Angel Network. If we combine “Undisclosed Investors” and “Undisclosed investors” (as they refer to the same category), the total becomes 69 investments, making them the clear leading investor group.


### Most Recently Started Startup's

```python
df['Datedd/mm/yyyy'] = pd.to_datetime(df['Datedd/mm/yyyy'], errors='coerce')
most=df.groupby('InvestorsName')['Datedd/mm/yyyy'].max().sort_values(ascending=False).head(5)
most
```

### Final Conclusion :

**Bengaluru** is the **leading city** for startups in India, supported by good infrastructure, skilled people, and strong investor support. Among investors, **Undisclosed Investors** lead with 69 investments, followed by Ratan Tata, Unknown, and Indian Angel Network. This shows that Bengaluru attracts both known and undisclosed investors, making it a key center for startup growth in India.
