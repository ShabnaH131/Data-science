```python
from collections import Counter
import ast
def extract_genre_names(genre_str):
    try:
        genres=ast.literal_eval(genre_str)
        return [genre['name']for genre in genres]
    except(SyntaxError,ValueError):
        return ['Unknown']

df['genre_list']=df['genres'].apply(extract_genre_names)
all_genres=[genre for sublist in df['genre_list'] for genre in sublist]
count_genres=Counter(all_genres)
genre_count_df=pd.DataFrame(count_genres.items(),columns=['genre','count'])
print("genres count")
for genre,count in genre_counts.items():
    print(f"{genre}: {count}")
    
```
